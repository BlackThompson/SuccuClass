```python
for folder_name in folder_names:
    wait = WebDriverWait(driver, 10)
    driver.get("https://www.bing.com/images/search?q=%E5%A4%9A%E8%82%89%E9%9B%AA%E8%8E%B2&form=BESBTB&first=1&cw=1177&ch=818")
    driver.find_element(By.ID,'sb_form_q').clear() #清除搜索框的内容
    driver.find_element(By.ID,'sb_form_q').send_keys("多肉" +folder_name+ Keys.RETURN) #搜索新的内容
    img_element=driver.find_element(By.CLASS_NAME,'mimg') #找到第一张缩略图
    img_element.click()
    
    for cnt in range(100):
        driver.switch_to.default_content()
        # 切换到指定id的iframe
        iframe = driver.find_element(By.ID,"OverlayIFrame")
        driver.switch_to.frame(iframe)
        img_element_large = driver.find_element(By.XPATH,"//div[@class='imgContainer']/img") #找到大图
        url_large=img_element_large.get_attribute('src')
        r=requests.get(url_large)

        # 目标文件夹路径
        destination_folder = '/Users/steafen/多肉/数据/'+folder_name+'/'
        filename=destination_folder+str(cnt)+'.jpg'
        with open(filename,'wb') as f:
            f.write(r.content)
        f.close
        
        wait = WebDriverWait(driver, 10)
        # 切换到下一张图片
        next_btn=driver.find_element(By.XPATH,"//div[@id='navr']/span[@class='icon']")
        next_btn.click()
        time.sleep(0.2)
        
    print(folder_name+" saved")
        
```

```python
driver = webdriver.Edge()
for folder_name in folder_names:
    wait = WebDriverWait(driver, 10)
    driver.get("https://www.bing.com/images/search?q=%E5%A4%9A%E8%82%89%E9%9B%AA%E8%8E%B2&form=BESBTB&first=1&cw=1177&ch=818")
    driver.find_element(By.ID,'sb_form_q').clear() #清除搜索框的内容
    driver.find_element(By.ID,'sb_form_q').send_keys("多肉植物" +folder_name+ Keys.RETURN) #搜索新的内容
    img_element=driver.find_element(By.CLASS_NAME,'mimg') #找到第一张缩略图
    img_element.click()
    
    for cnt in range(100):
        driver.switch_to.default_content()
        # 切换到指定id的iframe
        iframe = driver.find_element(By.ID,"OverlayIFrame")
        driver.switch_to.frame(iframe)
        
        while True:
            try:
                img_element_large = driver.find_element(By.XPATH,"//div[@class='imgContainer']/img") #找到大图
                url_large=img_element_large.get_attribute('src')
                r=requests.get(url_large)
                break
            except Exception as e:
                print("网络错误，等待并重试:", e)
                time.sleep(2)  # 等待5秒后重试

        # 目标文件夹路径
        destination_folder = '/Users/steafen/多肉/数据/'+folder_name+'/'
        filename=destination_folder+str(cnt)+'.jpg'
        with open(filename,'wb') as f:
            f.write(r.content)
        
        wait = WebDriverWait(driver, 10)
        # 切换到下一张图片
        next_btn=driver.find_element(By.XPATH,"//div[@id='navr']/span[@class='icon']")
        next_btn.click()
        time.sleep(0.2)
        
    print(folder_name+" saved")
```
